```python
import hashlib  #importar la libreria haslib con las funciones criptograficas hash

def H_text(mensaje): #funcion H_text que recibe un mensaje
    m = hashlib.sha256()  #crear objeto m que aplicalara el hahs 256
    m.update(mensaje.encode())  #los strings que recibe m los pasa a bytes usando el encode
    return m.hexdigest()   # devuelve el hash en formato hexadecimal
#definicion de hojas/textos
tx0 = "hola"
tx1 = "Hola"
tx2 = "chao"
tx3 = "Chao"
#aplicar hash-256 a cada texto y asi obtener hash de cada uno
h0 = H_text(tx0)
h1 = H_text(tx1)
h2 = H_text(tx2)
h3 = H_text(tx3)
#concatenar ramas
h01 = H_text(h0 + h1)
h23 = H_text(h2 +h3)

#root final
root = H_text(h01 + h23)
print(root)
```

```python
# 2 a la n entradas:
# modificado con chatgpt

import hashlib

def H_text(mensaje):
    m = hashlib.sha256()
    m.update(mensaje.encode())
    return m.hexdigest()

tx = ["hola", "Hola", "chao", "Chao"]

# Verificar que el tamaño sea potencia de 2
n = len(tx)
if n < 1 or (n & (n - 1)) != 0:
    raise ValueError("La cantidad de entradas debe ser potencia de 2 (2^n) y >= 1")

# Hash de hojas
nivel = [H_text(t) for t in tx]

# Construcción del árbol de abajo hacia arriba
while len(nivel) > 1:
    nivel = [H_text(nivel[i] + nivel[i+1]) for i in range(0, len(nivel), 2)]

root = nivel[0]
print(root)
```

```python
# n entradas mayores o iguales que 1, desconociendo el valor de n
# modificado con chatgpt

import hashlib

def H_text(mensaje):
    m = hashlib.sha256()
    m.update(mensaje.encode())
    return m.hexdigest()


tx = ["hola", "Hola", "chao", "Chao", "bye"]

if len(tx) < 1:
    raise ValueError("Se requiere al menos 1 entrada (n >= 1)")

# Hash de hojas
nivel = [H_text(t) for t in tx]

# Construcción del árbol (bottom-up) para n arbitrario
while len(nivel) > 1:
    if len(nivel) % 2 == 1:
        nivel.append(nivel[-1])  # duplicar el último si el nivel es impar

    nivel = [H_text(nivel[i] + nivel[i+1]) for i in range(0, len(nivel), 2)]

root = nivel[0]
print(root)
```
